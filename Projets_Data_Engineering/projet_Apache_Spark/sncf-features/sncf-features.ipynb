# import

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pyspark
from pyspark.sql import SparkSession
```

```python
# create SparkSession
spark = SparkSession.builder.master("local[*]") \
                    .appName('SparkSNCF') \
                    .getOrCreate()

# extract SparkContext
sc = spark.sparkContext

print('Spark is ready with CPU usage :', sc.defaultParallelism)
```

# import data

```python
table_prix = sc.textFile('tarifs-tgv-par-od.csv')
```

```python
#  table_prix.collect()
```

```python
table_prix_head = table_prix.take(1)
table_prix_no_header = table_prix.filter(lambda row: row != table_prix_head[0])
```

```python
# table_prix_no_header.collect()
```

```python
table_prix_splited = table_prix_no_header.map(lambda line: tuple(line.split(';')[0:4:3])) \
                                         .map(lambda x: tuple(x[0].split('-', 1) + [float(x[1])])) 
```

```python
table_prix_splited.collect()
```
